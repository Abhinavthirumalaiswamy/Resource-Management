```python
# Tamil nadu bar chart
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10,5))

manufacturers = ['furniture', 'dairy products', 'bakery products', 'wooden containers', 'cement']
Main_Workers = [1886, 4379, 1422, 7808, 2849]
bar_labels = ['red','blue','green','yellow','pink']
bar_colours = ['tab:red','tab:blue','tab:green','yellow','tab:pink']

for i in range(len(manufacturers)):
    ax.bar(manufacturers[i], Main_Workers[i], color=bar_colours[i], label=bar_labels[i])

ax.set_ylabel('manufacturers workers count')
ax.set_xlabel('products')
ax.set_title('Tamil nadu Manufactures supply of products')
ax.legend(title = 'Manufactures main workers counts')

plt.show()
```

```python
# Tamil nadu pie chart
import matplotlib.pyplot as plt

labels = ['furniture', 'dairy products', 'bakery products', 'wooden containers', 'cement']
Main_Workers = [1886, 4379, 1422, 7808, 2849]
colours = ['tab:red','tab:blue','tab:green','yellow','tab:pink']


plt.pie(Main_Workers, labels=labels, colors=colours)
plt.title('Tamil Nadu Manufactures supply of products')
plt.show()
```

```python
# Tamil nadu Donut chart
import matplotlib.pyplot as plt

labels = ['furniture', 'dairy products', 'bakery products', 'wooden containers', 'cement']
Main_Workers = [1886, 4379, 1422, 7808, 2849]
colours = ['tab:red','tab:blue','tab:green','yellow','tab:pink']
width_lister = [0.50]
for i in range(len(width_lister)):
  print(width_lister[i])
  plt.pie(Main_Workers, labels=labels, colors=colours, wedgeprops=dict(width=width_lister[i]), shadow = True, autopct = '%1.2f%%') # autopct = 1.2f    1 - one whole number 2 - precision values
  plt.title('Tamil Nadu Manufactures supply of products')
  plt.show()
```

```python
# Terminal command: pip install streamlit plotly pandas

import streamlit as st
import pandas as pd
import plotly.express as px

# 1. Required Data Definitions
data = {'Tamil Nadu': [1886, 4379, 1422, 7808, 2849], 'Kerala': [4231, 740, 1154, 2379, 1846], 'Karnataka': [8076, 3420, 1537, 13321, 2512], 'Maharastra': [16441, 3737, 2094, 9639, 2262], 'Gujarat': [1396, 1546, 431, 3193, 300]}
state_coords = {'Tamil Nadu': [11.12, 78.65], 'Kerala': [10.85, 76.27], 'Karnataka': [15.31, 75.71], 'Maharastra': [19.75, 75.71], 'Gujarat': [22.25, 71.19]}
labels = ['furniture', 'dairy products', 'bakery products', 'wooden containers', 'cement']

# 2. Data Processing
map_data_list = []
for state, counts in data.items():
    lat, lon = state_coords[state]
    for i, product in enumerate(labels):
        map_data_list.append({'State': state, 'Product': product, 'Workers': counts[i], 'lat': lat, 'lon': lon})
df_final = pd.DataFrame(map_data_list)

# 3. Visualizations
st.title("Worker Resource Maps 2026")
tab1, tab2 = st.tabs(["Distribution Map", "Intensity Heatmap"])

with tab1:
    fig_map = px.scatter_map(df_final, lat="lat", lon="lon", size="Workers", color="Product",
                             hover_name="State", zoom=3, height=600)
    st.plotly_chart(fig_map, use_container_width=True)

with tab2:
    fig_heat = px.density_map(df_final, lat="lat", lon="lon", z="Workers", radius=50,
                              center=dict(lat=20, lon=78), zoom=3, height=600)
    st.plotly_chart(fig_heat, use_container_width=True)
```
